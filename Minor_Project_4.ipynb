```
import random

ProductList = {'p1':10, 'p2':15, 'p3':20, 'p4':25, 'p5':30, 'p6':35, 'p7':50,
               'p8':40, 'p9':55, 'p10':60, 'p11':65, 'p12':75, 'p13':70,
               'p14':45}

LB = 50
UB = 150
ResultList = set()  # Store sets with product names and values
Iterations = 1000

for _ in range(Iterations):
    size = random.randint(1, len(ProductList) - 1)
    sample_items = random.sample(list(ProductList.items()), size)  # Convert items to list
    sample_dict = dict(sample_items)  # Convert sampled items back to a dictionary

    if LB <= sum(sample_dict.values()) <= UB:
        ResultList.add(tuple(sample_dict.items()))  # Store as a tuple for uniqueness

# Display results
for result in ResultList:
    print(result)


```
