```
import random
import string
```

```
##Q1
filefp=open("text.txt","w")
for i in range(0,1000):
  random_string = ''.join(random.choices(string.ascii_letters,k=10))
  #print(random_string)
  filefp.write(random_string+"\n")
filefp.close()
```

```
##Q2
#assuming one character is of one byte
#1MB=1*1024(kb)*1024(bytes)
target_size=5*1024*1024 #5mb
line_length=50 #each line wil have 50 characters
num_lines=target_size//(line_length+1)#+1 for new line character
print(num_lines)

filefp=open("5mb.txt","w")
for i in range(0,num_lines):
  random_string=''.join(random.choices(string.ascii_letters,k=50))
  filefp.write(random_string+"\n")
```

```
##Q3
def create5MB_file(number_of_files):
    for x in range(1,number_of_files+1):
        target_size=5*1024*1024 #5mb
        line_length=50 #each line wil have 50 characters
        num_lines=target_size//(line_length+1)#+1 for new line character
        print(num_lines)

        filefp=open(f"5mbFILE-({x}).txt","w")
        for i in range(0,num_lines):
          random_string=''.join(random.choices(string.ascii_letters,k=50))
          filefp.write(random_string+"\n")
    filefp.close()
create5MB_file(10)
```

```
#GB was too large I'll use MB instead
##Q4
def create_file(size_of_file_inmb):
    target_size=size_of_file_inmb*1024*1024 #GB
    line_length=50 #each line wil have 50 characters
    num_lines=target_size//(line_length+1)
    filefp=open(f"{size_of_file_inmb}mbFILE-x.txt","w")
    for i in range(0,num_lines):
        random_string=''.join(random.choices(string.ascii_letters,k=50))
        filefp.write(random_string+"\n")
    filefp.close()
create_file(100)
create_file(200)
create_file(300)
create_file(400)
create_file(500)
```

```
##Q5
for i in [100,200,300,400,500]:
  filefp=open(f"{i}mbFILE-x.txt","r")
  writefp=open(f"{i}mbwrite.txt","w")
  for X in filefp: #when you iterate over a file it automitaclly goes over line by line
    writefp.write(X.upper())
  filefp.close()
  writefp.close()
```

```
##Q6 (using multithreading)
import threading
def convert_file_to_upper(file_size):
  filefp=open(f"{file_size}mbFILE-x.txt","r")
  writefp=open(f"{file_size}mbwrite(MultiThreading).txt","w")
  for X in filefp: #when you iterate over a file it automitaclly goes over line by line
    writefp.write(X.upper())
  filefp.close()
  writefp.close()

file_sizes = [100, 200, 300, 400, 500]
threads =[]

for S in file_sizes:
  thread=threading.Thread(target=convert_file_to_upper,args=(S,))
  threads.append(thread)
  thread.start()

for x in threads:
  x.join()
print("Multithreading suscessful")
```

```
##Q12
import pandas as pd
import numpy as np
rows, cols = 100, 30

data = np.random.randint(1, 201, size=(rows, cols))#range 1,201
df = pd.DataFrame(data)
df.head()
```

```
df.replace(range(10, 61), np.nan, inplace=True)
df.head()
```

```
df.fillna(df.mean(), inplace=True)
df.head()
```

```
df = (df - df.min()) / (df.max() - df.min()) * 10
df.head()
```

```
df = df.applymap(lambda x: 1 if x <= 0.5 else 0)
df.head()
```
